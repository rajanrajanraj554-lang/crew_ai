## Step 3: Your First Prompt Template

A **prompt template** is like a form with blanks that you fill in. This ensures consistency.

```python
from langchain_core.prompts import PromptTemplate

# Create a template with a placeholder {topic}
template = PromptTemplate.from_template(
    "Explain {topic} in simple terms like I'm 5 years old."
)

# Fill in the placeholder
prompt = template.format(topic="gravity")
print("Generated Prompt:")
print(prompt)
```

## Step 4: Connect to an LLM

Now let's connect our template to a language model.

```python
from langchain_openai import ChatOpenAI

# Initialize the model
model = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.7)

print("Model initialized successfully!")
```

## Step 5: Add Output Parser

An **output parser** converts the AI's response into a usable format.

```python
from langchain_core.output_parsers import StrOutputParser

# Simple parser that returns text as-is
parser = StrOutputParser()

print("Parser created!")
```

## Step 6: Build Your First Chain

A **chain** connects components together: Prompt → Model → Parser

```python
# Create the chain using the | operator
chain = template | model | parser

# Run the chain!
result = chain.invoke({"topic": "photosynthesis"})

print("🤖 AI Response:")
print("=" * 50)
print(result)
```

### 🎯 Exercise 2
Create a chain that translates text to different languages.

```python
# YOUR CODE HERE
translate_template = PromptTemplate.from_template(
    "Translate '{text}' to {language}"
)

translate_chain = translate_template | model | parser

# Test it
translation = translate_chain.invoke({"text": "Hello, how are you?", "language": "French"})
print(f"Translation: {translation}")
```

## Step 7: JSON Output Parser

Sometimes you want structured data back. Let's use a JSON parser!

```python
from langchain_core.output_parsers import JsonOutputParser

# Create a JSON parser
json_parser = JsonOutputParser()

# Create a prompt that asks for JSON
json_prompt = PromptTemplate.from_template(
    """Provide information about {animal} in JSON format with these fields:
    - name: string
    - habitat: string
    - diet: string
    - lifespan: number (years)
    
    Only return valid JSON."""
)

# Create chain
json_chain = json_prompt | model | json_parser

# Get structured result
animal_info = json_chain.invoke({"animal": "elephant"})

print("Structured Data:")
print(animal_info)
print(f"\nType: {type(animal_info)}")
print(f"Name: {animal_info['name']}")
```

## Step 8: Multiple Variables Chain

Let's create a more complex example with multiple inputs.

```python
# Complex template with multiple variables
complex_template = PromptTemplate.from_template(
    """Write a {style} story about {character} who lives in {setting}.
    The story should be exactly {length} sentences long."""
)

story_chain = complex_template | model | parser

# Generate different stories
stories = [
    {"style": "funny", "character": "a cat", "setting": "a spaceship", "length": 3},
    {"style": "mysterious", "character": "a detective", "setting": "Victorian London", "length": 4},
    {"style": "inspirational", "character": "a young student", "setting": "a library", "length": 3}
]

for i, params in enumerate(stories, 1):
    print(f"\n{'='*50}")
    print(f"Story {i}: {params['style'].title()} Tale")
    print('='*50)
    story = story_chain.invoke(params)
    print(story)
```

## 🏆 Final Exercise

Create a "Study Helper" chain that:
1. Takes a subject and difficulty level
2. Generates explanation + quiz question + answer
3. Returns everything in JSON format

```python
# YOUR CODE HERE - Build your Study Helper!

study_prompt = PromptTemplate.from_template(
    """Create a study guide for {subject} at {difficulty} level.
    Return JSON with:
    - explanation: brief explanation of key concept
    - quiz_question: one question to test understanding
    - answer: correct answer
    - tip: study tip
    
    Only return valid JSON."""
)

study_chain = study_prompt | model | json_parser

# Test it
guide = study_chain.invoke({"subject": "Python loops", "difficulty": "beginner"})

print("📚 Study Guide:")
print(f"Topic: Python loops")
print(f"\n📖 Explanation:\n{guide['explanation']}")
print(f"\n❓ Quiz: {guide['quiz_question']}")
print(f"\n✅ Answer: {guide['answer']}")
print(f"\n💡 Tip: {guide['tip']}")
```
